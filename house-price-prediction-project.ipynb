```python
def show_rows(show=False):
    if show:
        pd.set_option("display.max_rows", None)
    else:
        pd.set_option("display.max_rows", 50)
#--------------------------------------------------------------------------------------------------------
def show_columns(show=False):
    if show:
        pd.set_option("display.max_columns", None)
    else:
        pd.set_option("display.max_columns", 20)
#--------------------------------------------------------------------------------------------------------
def dataframe_summary(dataframe, show_value_counts = False):
    import pandas as pd
    pd.set_option("display.colheader_justify","left")
    print("-------------------------------------------------------------------------------------------------------------------------------------------------")
    print("-------- SHAPE of Dataset -----------------------------------------------------------------------------------------------------------------------")
    print(dataframe.shape)
    print("-------------------------------------------------------------------------------------------------------------------------------------------------")
    print("-------- DATA TYPES of Dataset ------------------------------------------------------------------------------------------------------------------")
    print(dataframe.dtypes)
    print("-------------------------------------------------------------------------------------------------------------------------------------------------")
    print("-------- MEMORY USAGE of Dataset --------------------------------------------------------------------------------------------------------------  ")
    print(dataframe.memory_usage(deep=True) / (1024 * 1024))
    print("TOTAL >>> ", round((dataframe.memory_usage(deep=True) / (1024 * 1024)).sum(),2), "\n (All are in MB)")
    print("-------------------------------------------------------------------------------------------------------------------------------------------------")
    print("-------- MISSING VALUES in Dataset --------------------------------------------------------------------------------------------------------------")
    print(dataframe.isnull().sum().sort_values(ascending=False))
    print("-------------------------------------------------------------------------------------------------------------------------------------------------")
    print("-------- DESCRIPTIVE Info about Dataset ---------------------------------------------------------------------------------------------------------")
    print(dataframe.describe([0.01, 0.05, 0.25, 0.5, 0.75, 0.95, 0.99]).T)
    print("-------------------------------------------------------------------------------------------------------------------------------------------------")

    if show_value_counts:
        print("-------- VALUE COUNTS in Dataset ----------------------------------------------------------------------------------------------------------------")
        for col in dataframe.columns:
            print(pd.DataFrame({"Count": dataframe[col].value_counts(),
                               "Ratio": round((dataframe[col].value_counts() / len(df) * 100),2)})
                 )
            print("-------------------------------------------------------------------------------------------------------------------------------------------------")
#--------------------------------------------------------------------------------------------------------            
def grab_column_names_based_on_types(dataframe, cardinal_threshold_percent=0.6, category_threshold_num=20):
    def is_date(string):
        from datetime import datetime as dt
        try:
            if dt.strptime(string, '%Y-%m-%d %H:%M:%S'):
                return True
            else:
                return False
        except ValueError:
            return False
    def is_numeric(string):
        try:
            value_float = float(string)
            value_int = int(string)
            return True

        except ValueError:
            return False
    #cat_but_num_cols = [col for col in dataframe.columns if (dataframe[col].dtype in ["O"] and all(is_numeric(str(value)) for value in dataframe.dropna(subset=[col])[col]))]
    date_cols = [col for col in dataframe.columns if (dataframe[col].dtype in ["O"] and all(is_date(str(value)) for value in dataframe.dropna(subset=[col])[col]) or type(dataframe[col].iloc[0]) is pd.Timestamp)]
    num_but_car_cols = [col for col in dataframe.columns if dataframe[col].dtype in ["int", "float"] and dataframe[col].nunique() == (dataframe.shape[0])]
    cat_but_car_cols = [col for col in dataframe.columns if dataframe[col].dtype in ["O","category"] and dataframe[col].nunique() > (dataframe.shape[0] * cardinal_threshold_percent) and not col in date_cols]
    car_cols = num_but_car_cols + cat_but_car_cols
    num_but_cat_cols = [col for col in dataframe.columns if dataframe[col].dtype in ["int", "float"] and dataframe[col].nunique() < category_threshold_num]
    cat_cols = [col for col in dataframe.columns if dataframe[col].dtype in ["O","category","bool"] and not col in cat_but_car_cols and not col in date_cols] + num_but_cat_cols
    num_cols = [col for col in dataframe.columns if dataframe[col].dtype in ["int", "float"] and not col in num_but_cat_cols and not col in num_but_car_cols]

    print("------------------------------------------------------------------------------------------------------------")
    print(f"Observations Count: {dataframe.shape[0]}")
    print(f"Variables Count: {dataframe.shape[1]}")
    print("------------------------------------------------------------------------------------------------------------")
    print(f'Date Columns Count: {len(date_cols)} -> {date_cols}\n')
    print(f'Numeric Columns Count: {len(num_cols)} -> {num_cols}\n')
    print(f'Categorical Columns Count: {len(cat_cols)} -> {cat_cols}\n')
    print(f'Cardinal Columns Count: {len(car_cols)} -> {car_cols}')
    print("------------------------------------------------------------------------------------------------------------")
    print(f'Numeric but Cardinal Columns Count: {len(num_but_car_cols)} -> {num_but_car_cols}\n')
    print(f'Numeric but Categorical Columns Count: {len(num_but_cat_cols)} -> {num_but_cat_cols}\n')
    print(f'Categorical but Cardinal Columns Count: {len(num_but_cat_cols)} -> {cat_but_car_cols}\n')
    print("------------------------------------------------------------------------------------------------------------")

    if dataframe.shape[1] == (len(date_cols) + len(num_cols) + len(cat_cols) + len(car_cols)):
        print(f'Variables Count {dataframe.shape[1]} and Sum of Date,Numeric,Categorical and Cardinal Columns Counts {len(date_cols) + len(num_cols) + len(cat_cols) + len(car_cols)} are equal..')
    else:
        print(
            f'Variables Count {dataframe.shape[1]} and Sum of Date,Numeric,Categorical and Cardinal Columns Counts {len(date_cols) + len(num_cols) + len(cat_cols) + len(car_cols)} are NOT equal !!')
    print("------------------------------------------------------------------------------------------------------------")
     
    df_dtypes = pd.DataFrame(dataframe.dtypes).reset_index()

    def dtype_match(value):
        if value in num_cols:
            return "int/float"
        elif value in cat_cols:
            return "category"
        elif value in car_cols:
            return "cardinal"
        elif value in date_cols:
            return "datetime"

    def get_col_values(dataframe, column_name):
        return dataframe[column_name].unique()
    
    df_dtypes["dtype_indeed"] = df_dtypes["index"].apply(lambda x: dtype_match(x))
    df_dtypes["values"] = ""
    for inx, val in enumerate(df_dtypes["index"]):
        df_dtypes.at[inx, "values"] = list(df[val].unique())
        
    df_dtypes.columns = ["columns","dtype_is","dtype_may_be","unique_values"]
    df_dtypes.sort_values(by="dtype_may_be", inplace=True)
    print("----- Data Types Comparison --------------------------------------------------------------------------------")
    print("------------------------------------------------------------------------------------------------------------")
    pd.set_option("display.max_rows",None)
    pd.set_option("display.max_colwidth",100)
    #print(df_dtypes)
    display(df_dtypes)
    pd.reset_option('display.max_colwidth')
    pd.set_option("display.max_rows",50)
    print("------------------------------------------------------------------------------------------------------------")
    
    return num_cols, cat_cols, car_cols, date_cols
#--------------------------------------------------------------------------------------------------------            
def cat_summary(dataframe, column_name, plot=False):
    print(pd.DataFrame({column_name: dataframe[column_name].value_counts(),
                 "Ratio": 100 * (dataframe[column_name].value_counts() / len(dataframe))}))
    print("--------------------------------------------------------------------------------------------------------")

    if plot:
        sns.countplot(x=dataframe[column_name], data=dataframe)
        plt.show()
        print("--------------------------------------------------------------------------------------------------------")
#--------------------------------------------------------------------------------------------------------        
def num_summary(dataframe, column_name, plot = False, plot_bins = 20):

    quantiles = [0.05, 0.10, 0.20, 0.30, 0.40, 0.50, 0.60, 0.70, 0.80, 0.90, 0.95, 0.99]
    print(dataframe[column_name].describe(quantiles).T)
    print("--------------------------------------------------------------------------------------------------------")
    
    if plot:
        dataframe[column_name].hist(bins = plot_bins)
        plt.xlabel(column_name)
        plt.title(column_name)
        plt.show()
        print("--------------------------------------------------------------------------------------------------------")
#--------------------------------------------------------------------------------------------------------
def target_analysis(dataframe, target_label, columns, cat_analysis=True, num_analysis=True, show_boxplot=False):
    #- analysis of categorical variables
    if cat_analysis:
        print("---------------------------------------------")
        print("----- Analysis of Categorical Variables -----")
        print("---------------------------------------------")
        for col in [col for col in dataframe.columns if dataframe[col].dtype not in [int,float]]:
            print(f"----- {col} ---------------------------------------------------------------------")
            result = dataframe.groupby(col).agg({target_label: ["mean", "count"]}).round(2)
            result[("Ratios","count_in_total")] = (result[(target,"count")] / len(dataframe) * 100).round(2)
            result[("Ratios","sales_in_total")] = (result[(target,"mean")] * result[(target,"count")] / dataframe[target].sum() * 100).round(2)
            result.sort_values(by=[("Ratios","sales_in_total"),("Ratios","count_in_total")], ascending=False, inplace=True)
            display(result)
            print("----------------------------------------------------------------------------------")
            #categoricals_summary(df, col, plot=False)
        #---------------------------------------------------------------------------------------------
    #- analysis of numerical variables
    if num_analysis:
        print("---------------------------------------------")
        print("----- Analysis of Numerical Variables -------")
        print("---------------------------------------------")
        for col in [col for col in dataframe.columns if dataframe[col].dtype in [int,float]]:
            print(f"----- {col} ---------------------------------------------------------------------")
            display(dataframe.groupby(target_label).agg({col: ["mean", "min", "max"]}).round(2).sort_values(by=[(col,"mean"),(col,"max")], ascending=False))
            print("----------------------------------------------------------------------------------")
            
            if show_boxplot:
                print(f"----- {col} ---------------------------------------------------------------------")
                sns.boxplot(x=dataframe[col])
                plt.show(block=True)
                print("----------------------------------------------------------------------------------")
            
            #numericals_summary(df, col, plot=False)

def outlier_thresholds(dataframe, column_name, q1 = 0.25, q3 = 0.75, print_info=True):
    q1_num = dataframe[column_name].quantile(q1)
    q3_num = dataframe[column_name].quantile(q3)
    iqr = q3_num - q1_num
    lower_threshold = q1_num - iqr * 1.5
    upper_threshold = q3_num + iqr * 1.5
    if print_info:
        print(f"for {column_name}:")
        print(f"Q1 is {q1}")
        print(f"Q3 is {q3}")
        print(f"Q1 threshold is {q1_num}")
        print(f"Q3 threshold is {q3_num}")
        print(f"IQR is {iqr}")
        print(f"Lower threshold is {lower_threshold}")
        print(f"Upper Threshold is {upper_threshold}")
        print("-----------------------------------------------------------------------------------------------------------")
    return lower_threshold, upper_threshold
#----------------------------------------------------------------------------------------------------------------------
def check_if_outlier_exists(dataframe, column_name, q1 = 0.25, q3 = 0.75, print_details=True):
    low_limit, up_limit = outlier_thresholds(dataframe, column_name, q1, q3, print_details)
    if dataframe[((dataframe[column_name] < low_limit) | (dataframe[column_name] > up_limit))].shape[0] > 0:
        print(f"{column_name} : Outliers exist based on Q1={q1} and Q3={q3}")
        print(f"Under lower bound: {len(dataframe[dataframe[column_name] < low_limit])} observation of {len(dataframe)}")
        print(f"Above upper bound: {len(dataframe[dataframe[column_name] > up_limit])} observation of {len(dataframe)}")
        print("-------------------------------------------------------------------------------------------------------")
        return True
    else:
        print(f"{column_name} : No Outlier based on Q1={q1} and Q3={q3}")
        print("-------------------------------------------------------------------------------------------------------")
        return False
#----------------------------------------------------------------------------------------------------------------------
def grap_outliers(dataframe, column_name, q1 = 0.25, q3 = 0.75, get_incides=False, print_results=True):

    low_limit, up_limit = outlier_thresholds(dataframe, column_name, q1, q3)

    if print_results:
        if len(dataframe[(dataframe[column_name] < low_limit) | (dataframe[column_name] > up_limit)]) > 10:
            print(dataframe[(dataframe[column_name] < low_limit) | (dataframe[column_name] > up_limit)].head())
        else:
            print(dataframe[(dataframe[column_name] < low_limit) | (dataframe[column_name] > up_limit)])

    if get_incides:
        return dataframe[(dataframe[column_name] < low_limit) | (dataframe[column_name] > up_limit)].index
#----------------------------------------------------------------------------------------------------------------------
def rare_analyzer(dataframe, target_col_name, forCategoricals=True, forNumericals=False):
    if forCategoricals:
        for col in cat_cols:
            print(f'{col} Count: {len(dataframe[col])}')
            dff = pd.DataFrame({"Count": dataframe[col].value_counts(),
                                "Ratio": dataframe[col].value_counts() / len(dataframe),
                                "Target_Mean": dataframe.groupby(col)[target_col_name].mean()})
            print(dff.sort_values(by=["Ratio", "Target_Mean", "Count"], ascending=[False, False, False]))
            print("-----------------------------------------------------------------------------------")

    if forNumericals:
        for col in num_cols:
            print(f'{col} Count: {len(dataframe[col])}')
            dff = pd.DataFrame({"Count": dataframe[col].value_counts(),
                                "Ratio": dataframe[col].value_counts() / len(dataframe),
                                "Target_Mean": dataframe.groupby(col)[target_col_name].mean()})
            print(dff.sort_values(by=["Ratio", "Target_Mean", "Count"], ascending=[False, False, False]))
            print("-----------------------------------------------------------------------------------")
#----------------------------------------------------------------------------------------------------------------------
def local_outlier_factor_analysis(dataframe, plot=False, return_scores= False, n_neighbors=20):
    lof = LocalOutlierFactor(n_neighbors=n_neighbors)
    lof.fit_predict(dataframe)
    scores_neg = np.sort(lof.negative_outlier_factor_)
    if plot:
        pd.DataFrame(scores_neg).plot(stacked=True, xlim = [0,30], style=".-")
        plt.show()
    if return_scores:
        return scores_neg

#----------------------------------------------------------------------------------------------------------------------            
def corr_matrix_for_num_cols(dataframe, numeric_columns, show_plot=True, return_corr_matrix=False):
    print("-------- Correlation Mapping ---------")
    print("Very Weak Correlation:   (0.00 - 0.19)")
    print("Weak Correlation:        (0.20 - 0.39)")
    print("Moderate Correlation:    (0.40 - 0.59)")
    print("Strong Correlation:      (0.60 - 0.79)")
    print("Very Strong Correlation: (0.80 - 1.00)")
    print("--------------------------------------")
    corr_matrix = dataframe[numeric_columns].corr()
    if show_plot:
        size_plt = len(corr_matrix) * 0.8
        plt.figure(figsize=(size_plt,size_plt))
        sns.heatmap(round(corr_matrix,2), vmin=-1, vmax=1, annot=True, cmap="RdBu")
        plt.show(block=True)
    if return_corr_matrix:
        return corr_matrix
#----------------------------------------------------------------------------------------------------------------------
def missing_values_summary(dataframe, print_df=True, return_na_col_names=False,return_na_df=False):
    na_cols = [col for col in dataframe.columns if dataframe[col].isnull().sum() > 0]
    na_list = dataframe[na_cols].isnull().sum().sort_values(ascending=False)
    na_ratios = (dataframe[na_cols].isnull().sum() / dataframe.shape[0] * 100).sort_values(ascending=False)
    na_df = pd.concat([na_list,na_ratios.round(2)], axis=1, keys=["na_counts","na_ratios"])
    #na_df = pd.DataFrame(na_ratios).reset_index().rename(columns={"index": "NA_Column", 0: "Ratio"})
    if print_df:
        if len(na_df) == 0:
            print("There is no missing values in the dataset.")
        else:
            display(na_df)

    if return_na_col_names:
        return na_cols

    if return_na_df:
        return na_df
#----------------------------------------------------------------------------------------------------------------------
def missing_corr_analysis(dataframe, display_corr=False, plot_show=True):
    temp_df = dataframe.copy()
    na_columns = missing_values_summary(temp_df, return_na_col_names=True, print_df=False)
    for col in na_columns:
        temp_df[col + "_na"] = np.where(temp_df[col].isnull(), 1, 0)
    na_flags = temp_df.loc[:, temp_df.columns.str.contains("_na")].columns
    corr_missing = temp_df[na_flags].corr()
    if display_corr:
        display(corr_missing)
    if plot_show:
        size_plt = len(corr_missing) * 0.7
        plt.figure(figsize=(size_plt,size_plt))
        sns.heatmap(corr_missing.round(2), cmap="RdBu", annot=True, vmax=1, vmin=-1)
        plt.show(block=True)
#----------------------------------------------------------------------------------------------------------------------
def missing_analysis(dataframe, target_col, show_na_values=True, show_na_vs_target=True, show_corr_plot=True):
    
    if show_na_values:
        print("---------------------------------------------")
        print("------------- Missing Values ----------------")
        print("---------------------------------------------")

        temp_df = dataframe.copy()
        na_columns = missing_values_summary(temp_df, return_na_col_names=True)
    
    if show_na_vs_target:
        print("---------------------------------------------")
        print("----- Missing Values vs. Target Analysis ----")
        print("---------------------------------------------")

        for col in na_columns:
            temp_df[col + "_na_flag"] = np.where(temp_df[col].isnull(), 1, 0)

        na_flags = temp_df.loc[:, temp_df.columns.str.contains("_na_flag")].columns

        for col in na_flags:
            display(pd.DataFrame({"Target_Mean": temp_df.groupby(col)[target_col].mean(),
                                "Count": temp_df.groupby(col)[target_col].count()}))
            print("----------------------------------------------------------------")
    
    if show_corr_plot:
        print("---------------------------------------------")
        print("---- Missing Values Correlation Analysis ----")
        print("---------------------------------------------")
        missing_corr_analysis(dataframe, plot_show=True)
        print("----------------------------------------------------------------")
#----------------------------------------------------------------------------------------------------------------------

#----------------------------------------------------------------------------------------------------------------------
def remove_outliers(dataframe, col_name, q1=0.25, q3=0.75):
    low_limit, up_limit = outlier_thresholds(dataframe, col_name, q1, q3)
    df_without_outliers = dataframe[~((df[col_name] < low_limit) | (dataframe[col_name] > up_limit))]
    return df_without_outliers
#----------------------------------------------------------------------------------------------------------------------
def winsorize_with_thresholds(dataframe, col_name, q1=0.25, q3=0.75):
    low_limit, up_limit = outlier_thresholds(dataframe, col_name, q1, q3)
    dataframe.loc[(dataframe[col_name] < low_limit), col_name] = low_limit
    dataframe.loc[(dataframe[col_name] > up_limit), col_name] = up_limit
#----------------------------------------------------------------------------------------------------------------------
def local_outlier_factor_analysis(dataframe, num_cols, plot=False, return_scores= False, n_neighbors=20):
    lof = LocalOutlierFactor(n_neighbors=n_neighbors)
    lof.fit_predict(dataframe[num_cols].dropna())
    scores_neg = np.sort(lof.negative_outlier_factor_)
    if plot:
        pd.DataFrame(scores_neg).plot(stacked=True, xlim = [0,30], style=".-")
        plt.show()
    if return_scores:
        return scores_neg
#----------------------------------------------------------------------------------------------------------------------
def rare_encoder(dataframe, rare_perc):
    temp_df = dataframe. copy()
    rare_columns = [col for col in temp_df.columns if temp_df[col].dtypes in ['O',"category"] and ((temp_df[col].value_counts() / len(temp_df)) < rare_perc).any(axis=None)]
    for var in rare_columns:
        tmp = temp_df[var].value_counts() / len(temp_df)
        rare_labels = tmp[tmp < rare_perc].index
        temp_df[var] = np.where(temp_df[var].isin(rare_labels), 'Rare', temp_df[var])

    return temp_df
#----------------------------------------------------------------------------------------------------------------------
def label_encoder_binary(dataframe, return_col_names=True, return_encoded_dataframe=False):
    if return_encoded_dataframe:
        return_col_names = False
    df_ = dataframe.copy()
    binary_columns = []
    label_encoder = LabelEncoder()
    binary_columns = [col for col in df_.columns if df_[col].dtype not in [int, float] and df_[col].nunique() == 2]
    if len(binary_columns) > 0:
        for col in binary_columns:
            df_[col] = label_encoder.fit_transform(df_[col])
    if return_col_names:
        return binary_columns
    if return_encoded_dataframe:
        return df_
#-----------------------------------------------------------------------------------------------------------------------
def one_hot_encoder(dataframe, categorical_columns, drop_first=True, return_dataframe=False, print_df=False):
    df_ = dataframe.copy()
    df_ = pd.get_dummies(df_, columns=categorical_columns, drop_first=drop_first)
    
    if print_df:
        print(df_.head())
    if return_dataframe:
        return df_
#-----------------------------------------------------------------------------------------------------------------------    
def feature_importance_levels (model, features, lenght, save=False):
    feature_imp = pd.DataFrame({'Value': model.feature_importances_, 'Feature': features.columns})
    plt.figure(figsize=(10, 10))
    sns.set(font_scale=1)
    sns.barplot(x="Value", y="Feature", data=feature_imp.sort_values(by="Value",ascending=False)[0: lenght])
    plt.title('Features') 
    plt.tight_layout()
    plt.show()
```

## Splitting the Train and Test data. 

---
Data with empty values in the SalePrice variable is considered as the test set.

```python
#- split train test datasets
df_train = df[~df[target].isnull()]
df_test = df[df[target].isnull()]
```

## Building models using the training data and evaluate its performance. 
Also, building models with log transformation of the target variable and observe the results.
---

### >> Dataset Preparation

```python
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV, cross_validate, train_test_split
from sklearn.metrics import mean_squared_error, r2_score

#- split train test data on df_train
X = df_train.drop(target,axis=1)
y = df_train[target]

#- train_test_split of df_train
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3)
```

### >> LightGBMRegressor Model

```python
from lightgbm import LGBMRegressor

#- build model
lgb = LGBMRegressor(random_state=42).fit(X_train,y_train)

cv_lgb = cross_validate(lgb, X_train, y_train, cv=10,scoring=["r2","neg_mean_squared_error"],n_jobs=-1,verbose=True)

mse_train_lgb_base = abs(cv_lgb['test_neg_mean_squared_error'].mean())
r2_train_lgb_base = abs(cv_lgb['test_r2'].mean())

print("----- TRAIN RESULTS ----")
print("-"*50)
print(f"MSE of 10-fold CV: {mse_train_lgb_base:.2f}")
print("-"*50)
print(f"r2 of 10-fold CV: {r2_train_lgb_base:.4f}")
print("-"*50)

y_pred = lgb.predict(X_test)

mse_test_lgb_base = mean_squared_error(y_test,y_pred)
r2_test_lgb_base = r2_score(y_test,y_pred)

print("----- TEST RESULTS ----")
print("-"*50)
print(f"MSE of 10-fold CV: {mse_test_lgb_base:.2f}")
print("-"*50)
print(f"r2 of 10-fold CV: {r2_test_lgb_base:.4f}")
print("-"*50)
```

### >> LightGBMRegressor Model (with log transformation)

```python
# log transformation of y_train
y_train_log = np.log1p(y_train)
y_test_log = np.log1p(y_test)

#- build model
lgb_log = LGBMRegressor(random_state=42).fit(X_train,y_train_log)

cv_lgb = cross_validate(lgb_log, X_train, y_train_log, cv=10,scoring=["r2","neg_mean_squared_error"],n_jobs=-1,verbose=True)

mse_train_lgb_log = abs(cv_lgb['test_neg_mean_squared_error'].mean())
r2_train_lgb_log = abs(cv_lgb['test_r2'].mean())

print("----- TRAIN RESULTS (log) ----")
print("-"*50)
print(f"MSE of 10-fold CV: {mse_train_lgb_log:.2f}")
print("-"*50)
print(f"r2 of 10-fold CV: {r2_train_lgb_log:.4f}")
print("-"*50)

y_pred_log = lgb_log.predict(X_test)

mse_test_lgb_log = mean_squared_error(y_test_log,y_pred_log)
r2_test_lgb_log = r2_score(y_test_log,y_pred_log)

print("----- TEST RESULTS (log) ----")
print("-"*50)
print(f"MSE of 10-fold CV: {mse_test_lgb_log:.2f}")
print("-"*50)
print(f"r2 of 10-fold CV: {r2_test_lgb_log:.4f}")
print("-"*50)
```

### >> CatBoostRegressor Model

```python
from catboost import CatBoostRegressor

#- build model
cat = CatBoostRegressor(verbose=False,random_state=42).fit(X_train,y_train)

cv_cat = cross_validate(cat, X_train, y_train, cv=10, scoring=["r2","neg_mean_squared_error"], n_jobs=-1, verbose=True)

mse_train_cat_base = abs(cv_cat["test_neg_mean_squared_error"].mean())
r2_train_cat_base = abs(cv_cat["test_r2"].mean())

print("----- TRAIN RESULTS ----")
print("-"*50)
print(f"MSE of 10-fold CV: {mse_train_cat_base:.2f}")
print("-"*50)
print(f"r2 of 10-fold CV: {r2_train_cat_base:.4f}")
print("-"*50)

y_pred = cat.predict(X_test)
mse_test_cat_base = mean_squared_error(y_test,y_pred)
r2_test_cat_base = r2_score(y_test,y_pred)

print("----- TEST RESULTS ----")
print("-"*50)
print(f"MSE of 10-fold CV: {mse_test_cat_base:.2f}")
print("-"*50)
print(f"r2 of 10-fold CV: {r2_test_cat_base:.4f}")
print("-"*50)
```

### >> CatBoostRegressor Model (with log transformation)

```python
# log transformation of y_train
y_train_log = np.log1p(y_train)
y_test_log = np.log1p(y_test)

#- build model
cat_log = CatBoostRegressor(verbose=False,random_state=42).fit(X_train,y_train_log)

cv_cat = cross_validate(cat_log, X_train, y_train_log, cv=10, scoring=["r2","neg_mean_squared_error"], n_jobs=-1, verbose=True)

mse_train_cat_log = abs(cv_cat["test_neg_mean_squared_error"].mean())
r2_train_cat_log = abs(cv_cat["test_r2"].mean())

print("----- TRAIN RESULTS (log) ----")
print("-"*50)
print(f"MSE of 10-fold CV: {mse_train_cat_log:.2f}")
print("-"*50)
print(f"r2 of 10-fold CV: {r2_train_cat_log:.4f}")
print("-"*50)

y_pred_log = cat_log.predict(X_test)

mse_test_cat_log = mean_squared_error(y_test_log,y_pred_log)
r2_test_cat_log = r2_score(y_test_log,y_pred_log)

print("----- TEST RESULTS (log) ----")
print("-"*50)
print(f"MSE of 10-fold CV: {mse_test_cat_log:.2f}")
print("-"*50)
print(f"r2 of 10-fold CV: {r2_test_cat_log:.4f}")
print("-"*50)
```

## Perform hyperparameter optimization.
---

### >> LightGBMRegressor Hyperparameter Optimisation

```python
lgb.get_params()
```

```python
# find best parameters
params = {'learning_rate': np.arange(0.05,0.15,0.01),        
             'max_depth': [5,10,15],                         
             'min_child_samples': range(10,30,5),            
             'n_estimators': range(100,500,100)              
             }

lgb_rand = RandomizedSearchCV(lgb,
                             params,
                             n_iter=50,
                             cv=5,
                             scoring="neg_mean_squared_error",
                             n_jobs=-1,
                             verbose=True).fit(X_train,y_train)

lgb_rand.best_params_
```

```python
# final lgb model
lgb_final = LGBMRegressor(random_state=42)
lgb_final.set_params(**lgb_rand.best_params_)
lgb_final.fit(X_train, y_train)

cv_lgb = cross_validate(lgb_final, 
                        X_train, 
                        y_train, 
                        cv=10, 
                        scoring=["r2","neg_mean_squared_error"],
                        n_jobs=-1,
                        verbose=True)

mse_train_lgb_final = abs(cv_lgb['test_neg_mean_squared_error'].mean())
r2_train_lgb_final = abs(cv_lgb['test_r2'].mean())

print("----- TRAIN RESULTS ----")
print("-"*50)
print(f"MSE of 10-fold CV: {mse_train_lgb_final:.2f}")
print("-"*50)
print(f"r2 of 10-fold CV: {r2_train_lgb_final:.4f}")
print("-"*50)

y_pred = lgb_final.predict(X_test)

mse_test_lgb_final = mean_squared_error(y_test,y_pred)
r2_test_lgb_final = r2_score(y_test,y_pred)

print("----- TEST RESULTS ----")
print("-"*50)
print(f"MSE of 10-fold CV: {mse_test_lgb_final:.2f}")
print("-"*50)
print(f"r2 of 10-fold CV: {r2_test_lgb_final:.4f}")
print("-"*50)
```

#### *Function for plotting feature importance...*

```python
def plot_importance(model, dataframe, num=len(X), save=False):
    feature_imp = pd.DataFrame({'Value': model.feature_importances_,'Feature': dataframe.columns})
    plt.figure(figsize=(10, 10))
    sns.set(font_scale=1)
    sns.barplot(x="Value", y="Feature", data=feature_imp.sort_values(by="Value", ascending=False) [0:num])
    plt.title('Features')
    plt.tight_layout()
    plt.show()
    if save:
        plt.savefig('importances.png')
```
